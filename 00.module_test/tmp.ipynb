```python
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 6, 1)
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

```python
x = np.arange(0, 6, 0.1)
y = np.sin(x)
z = np.cos(x)

plt.plot(x,y,label="sin")
plt.plot(x,z,label="cos", linestyle="--")
plt.title("test")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()

plt.show()
```
