```python
import numpy as np
```

```python
x = np.array([1.0, 2.2, 3.0, 2])
y = np.array([0.1])

print(x * y)
```

```python
a = np.array([[1.0, 2.0], [3.0, 4.0], [3.0, 4.0]])
print((a+a).shape)
print(a+a)
print(a * 10)
print(a.flatten())
print(a > 10)
print(a[a > 2])
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 6, 1)
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

```python
x = np.arange(0, 6, 0.1)
y = np.sin(x)
z = np.cos(x)

plt.plot(x,y,label="sin")
plt.plot(x,z,label="cos", linestyle="--")
plt.title("test")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()

plt.show()
```
